# count survive and unsurvive patients

```python
df.head(5)
```

```python
var1=df.Survived.value_counts()
```

```python
plt.pie(var1, shadow=True,autopct='%0.2f%%',labels=["Survived","Not Survived"],colors=['silver','brown'])
plt.show()
```

# find how many female passenger had travelled in first class and show in pie chart

```python
df.head(5)
```

```python
female=df['Sex'].value_counts()
female

```

```python
df_female=df[df.Sex =="female"]
```

```python
var2=df_female.Pclass.value_counts()
var2
```

```python
plt.pie(var2,labels=["third class","first class","second class"],autopct='%0.2f%%',explode=(0,0.07,0))
plt.title("Female Passenger Travelled IN Different Classes")
plt.show()
```

# Find out how many female passenger had survived and their age < 30 ,show on pie chart and bar graph

```python
var3=df_female[df_female.Age<30]
```

```python
var4=df[(df.Age<30) & (df.Sex=="female")]
```

```python
var4=var4.Survived.value_counts()
```

```python
plt.pie(var4,labels=["survived","not survived"],autopct='%0.2f%%')
plt.title("Survived female passenger with age < 30")
plt.show()
```

```python
labels=["survived","not survived"]
plt.bar(labels,var4,width=0.5)
plt.title("Survival vs Non-Survival")
plt.ylabel('Number of Passengers')
plt.xlabel('Survival Status')
plt.title('Survival vs Non-Survival')
plt.show()
```

# show age with 20 bins

```python
age=df.Age
age
```

```python
counts, bins, _ =plt.hist(age, bins=20, color='#f0ad9e',edgecolor='black',label='Hist')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.plot(bin_centers,counts,marker='o',label='Age Distribution')
plt.xticks(np.arange(0,90,5))
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.legend()
plt.show()
```

```python
var6=df[df.Survived==1].Age
```

```python
counts, bins, _ =plt.hist(var6, bins=20, color='#f0ad9e',edgecolor='black',label='Age BIns')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.plot(bin_centers,counts,marker='o',label='Survived Age Distribution')
plt.xticks(np.arange(0,90,5))
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Survived Age Distribution")
plt.legend()
plt.show()
```

```python
var7=df[df.Survived==0].Age
```

```python
counts, bins, _ =plt.hist(var7, bins=20, color='#f0ad9e',edgecolor='black',label='Age BIns')
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.plot(bin_centers,counts,marker='o',label='Not Survived Age Distribution')
plt.xticks(np.arange(0,90,5))
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Not Survived Age Distribution")
plt.legend()
plt.show()
```
